```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
X = np.random.rand(100,1)*10
true_slope = 2.5
true_intercept = 5
noise = np.random.normal(0,2,size=(100,1))
Y = true_slope * X +true_intercept + noise
def predict(theta_0, theta_1, X):
    return theta_0+theta_1*X
def cost(theta_0,theta_1,X,Y):
    predictions = predict(theta_0,theta_1,X)
    errors = predictions - Y
    squared_errors = errors**2
    return 1/(2*len(X)) * np.sum(squared_errors)
def gradient_descent(X,Y,theta_0,theta_1,learning_rate,num_iterations):
    m=len(X)
    cost_history = []
    for i in range(num_iterations):
        predictions = predict(theta_0,theta_1,X)
        errors = predictions - Y
        gradient_0 = (1/m)*np.sum(errors)
        gradient_1 = (1/m)*np.sum(errors*X)
        theta_0 -= learning_rate*gradient_0
        theta_1 -= learning_rate*gradient_1
        cost_history.append(cost(theta_0,theta_1,X,Y))
    return theta_0,theta_1,cost_history
theta_0 = 0
theta_1 = 1
learning_rate = 0.01
num_iterations = 1000
theta_0,theta_1,cost_history=gradient_descent(X,Y,theta_0,theta_1,learning_rate, num_iterations)
plt.figure()
plt.plot(range(num_iterations),cost_history)
plt.xlabel("Number of Iterations")
plt.ylabel("Cost")
plt.title("Cost Curve")
plt.show()

print("Optimization Steps: ")
print("Iteration Cost ")
for i in range(num_iterations):
    print(f"{i}\t\t{cost_history[i]}")
learning_rate = [0.01,0.1,0.5]
plt.figure(figsize=(10,6))
for lr in learning_rate:
    theta_0,theta_1,cost_history=gradient_descent(X,Y,theta_0,theta_1,lr, num_iterations)
    plt.plot(range(num_iterations),cost_history,label=f'lr = {lr}')
plt.xlabel('Iterations')
plt.ylabel('cost')
plt.title('Effect of learning rate on optimization')
plt.legend()
plt.show()
    
```
